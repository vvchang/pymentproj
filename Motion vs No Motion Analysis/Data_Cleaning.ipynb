```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_headmotion=pd.read_csv('motion_predictions.csv')
df_noheadmotion=pd.read_csv('nomotion_predictions.csv')
```

```python
df_participantdata = pd.read_csv('participants.tsv', sep='\t')
```

```python
df_participantdata.head()
```

```python
df_participantdata = df_participantdata.drop(
    df_participantdata[
        df_participantdata['participant_id'].isin(['sub-NC225', 'sub-NC251'])
    ].index
)
```

```python
df_participantdata['real_age']=df_participantdata['age']
df_participantdata['real_sex']=df_participantdata['sex']
df_participantdata=df_participantdata.drop('sex', axis=1)
df_participantdata=df_participantdata.drop('age', axis=1)
```

```python
df_participantdata.head()
```

```python
df_noheadmotion['participant_id']='sub-'+df_noheadmotion['subject']
df_headmotion['participant_id']='sub-'+df_headmotion['subject']
```

```python
df_noheadmotion=df_noheadmotion.drop(['source', 'run', 'session', 'subject'],axis=1)
df_headmotion=df_headmotion.drop(['source', 'run', 'session', 'subject'],axis=1)
```

```python
df_noheadmotion = df_noheadmotion.rename(
    columns=lambda col: col if col == 'participant_id' else 'nomotion_' + col
)

df_headmotion = df_headmotion.rename(
    columns=lambda col: col if col == 'participant_id' else 'motion_' + col
)
```

```python
df= pd.merge(df_participantdata, df_noheadmotion, on='participant_id')
df= pd.merge(df, df_headmotion, on='participant_id')
```

```python
df.head()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
